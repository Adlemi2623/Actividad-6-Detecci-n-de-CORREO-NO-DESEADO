```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

# 20 SPAM y 20 HAM
mensajes = [

# SPAM
"gana dinero rapido desde casa",
"oferta especial compra ahora",
"has ganado un premio",
"haz clic para recibir regalo",
"gana premios registrandote",
"oferta limitada solo hoy",
"credito inmediato disponible",
"gana dinero en internet",
"ingresos desde tu celular",
"promocion exclusiva para ti",
"compra ahora con descuento",
"seleccionado para ganar premio",
"descarga app y gana dinero",
"regalo gratis por registrarte",
"invierte hoy y gana dinero",
"haz clic para recompensa",
"premio especial disponible",
"oferta unica no la pierdas",
"ganancias rapidas garantizadas",
"multiplica tu dinero",

# HAM
"hola como estas",
"nos vemos mañana",
"recuerda hacer la tarea",
"la reunion es a las tres",
"te envio el archivo",
"vamos a estudiar",
"avisa cuando llegues",
"hoy tenemos examen",
"el proyecto es mañana",
"gracias por tu ayuda",
"nos vemos en biblioteca",
"voy camino a casa",
"te llamo despues",
"el clima esta agradable",
"terminaste la actividad",
"vamos al cine",
"necesito tu opinion",
"ya envie el correo",
"nos vemos despues",
"trae los documentos"
]

labels = ["spam"]*20 + ["ham"]*20

df = pd.DataFrame({"mensaje":mensajes,"tipo":labels})

# limpiar texto
def limpiar(texto):
    texto = texto.lower()
    texto = re.sub(r"[^a-záéíóúñ\s]","",texto)
    return texto

df["mensaje"] = df["mensaje"].apply(limpiar)

# convertir texto a numeros
vector = TfidfVectorizer()
X = vector.fit_transform(df["mensaje"])

# entrenar modelo
modelo = MultinomialNB()
modelo.fit(X, df["tipo"])

# probar mensaje
mensaje = input("Escribe un mensaje: ")

mensaje = limpiar(mensaje)

pred = modelo.predict(vector.transform([mensaje]))

print("Resultado:", pred[0].upper())
```
